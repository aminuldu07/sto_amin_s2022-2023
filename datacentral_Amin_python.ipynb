```python
import pandas as pd
import numpy as np

# Replace 'path/to/file.csv' with the actual file path of your CSV file
df = pd.read_csv('C:/Users/mdaminulisla.prodhan/All_My_Miscellenous/Datacentral/no_zero_na_df_1532.csv')

# Check if the dataframe has any null or NaN values
if df.isnull().values.any():
    print("The dataframe contains null or NaN values.")
else:
    print("The dataframe does not contain null or NaN values.")

# Check if the dataframe has any zero or empty values
if (df == 0).values.any() or (df == '').values.any():
    print("The dataframe contains zero or empty values.")
else:
    print("The dataframe does not contain zero or empty values.")
    
# Count the number of zero and empty string values in each column
zero_counts = (df == 0).sum(axis=0)
empty_counts = (df == '').sum(axis=0)

# Print the results
print("Columns with zero values:\n", zero_counts[zero_counts > 0])
print("Columns with empty string values:\n", empty_counts[empty_counts > 0])


# Create a copy of the original dataframe
new_df = df.copy()

# Replace all zero values with NaN
new_df.replace(0, np.nan, inplace=True)

# Drop any rows with NaN values
new_df.dropna(inplace=True)

print(new_df)

# Write the new dataframe to a CSV file
#new_df.to_csv('no_zero_na_df_1532_py_1522row.csv', index=False)
```
