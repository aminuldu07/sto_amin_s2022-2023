```python
import pandas as pd

# Example data frame
df = pd.DataFrame({
    'STUDY_ID': [1, 1, 2, 2, 3, 3,2],
    'SET_CODE': ['A', 'B', 'A', 'B', 'A', 'B','B'],
    'SHIT': [10, 20, 30, 40, 50, 60,100],
    'EXTRA_COL': ['X', 'Y', 'Z', 'X', 'Y', 'Z','Q']
})
```

```python
df
```

```python
# Get unique combinations of STUDY_ID and SET_CODE
unique_combinations = df[['STUDY_ID', 'SET_CODE']].drop_duplicates()
unique_combinations
```

```python
# Create empty data frame to store results
result = pd.DataFrame(columns=['STUDY_ID', 'SET_CODE', 'SHIT_mean', 'SHIT_std'])
```

```python
result
```

```python
i=5
print("STUDY ID IS:",unique_combinations.iloc[i]['STUDY_ID'])
print("SET CODE IS:",unique_combinations.iloc[i]['SET_CODE'])
subset_df = df[(df['STUDY_ID'] == unique_combinations.iloc[i]['STUDY_ID']) & 
                   (df['SET_CODE'] == unique_combinations.iloc[i]['SET_CODE'])][['SHIT']]
print(subset_df)
```

```python
# Loop through unique combinations and calculate mean and std of SHIT
for i in range(len(unique_combinations)):
    subset_df = df[(df['STUDY_ID'] == unique_combinations.iloc[i]['STUDY_ID']) & 
                   (df['SET_CODE'] == unique_combinations.iloc[i]['SET_CODE'])][['SHIT']]
    mean_SHIT = subset_df.mean()[0]
    std_SHIT = subset_df.std()[0]
    result = result.append({'STUDY_ID': unique_combinations.iloc[i]['STUDY_ID'], 
                            'SET_CODE': unique_combinations.iloc[i]['SET_CODE'],
                            'SHIT_mean': mean_SHIT, 'SHIT_std': std_SHIT}, 
                           ignore_index=True)
```

```python
print(result)
```
