```python
import pandas as pd

# Example data frame
df = pd.read_csv('NCBW.csv')
```

```python
# Get unique combinations of STUDY_ID and SET_CODE
unique_combinations = df[['STUDYID', 'SETCD']].drop_duplicates()
unique_combinations'lhg
```

```python
# Create empty data frame to store results
result = pd.DataFrame(columns=['STUDYID', 'SETCD', 'BWSTRESN_mean', 'BWSTRESN_std'])
```

```python
result
```

```python
i=5
print("STUDYID IS:",unique_combinations.iloc[i]['STUDYID'])
print("SETCD IS:",unique_combinations.iloc[i]['SETCD'])
subset_df = df[(df['STUDYID'] == unique_combinations.iloc[i]['STUDYID']) & 
                   (df['SETCD'] == unique_combinations.iloc[i]['SETCD'])][['BWSTRESN']]
print(subset_df)
```

```python
# Loop through unique combinations and calculate mean and std of SHIT
for i in range(len(unique_combinations)):
    subset_df = df[(df['STUDYID'] == unique_combinations.iloc[i]['STUDYID']) & 
                   (df['SETCD'] == unique_combinations.iloc[i]['SETCD'])][['BWSTRESN']]
    mean_BWSTRESN = subset_df.mean()[0]
    std_BWSTRESN = subset_df.std()[0]
    result = result.append({'STUDYID': unique_combinations.iloc[i]['STUDYID'], 
                            'SETCD': unique_combinations.iloc[i]['SETCD'],
                            'BWSTRESN_mean': mean_BWSTRESN, 'BWSTRESN_std': std_BWSTRESN}, 
                           ignore_index=True)
```
